```python
import pandas as pd
```

```python
# inserting data set
```

```python
df = pd.read_csv('store_sku_ba_dataset.csv')
```

```python
df.head()
```

# TASK 1

```python
# Grouping by StoreID
```

```python
grp_StoreID = df.groupby('Store ID')[['Total Visits', 'Transactions']].sum()
grp_StoreID['conversion rate Store level'] = grp_StoreID['Transactions'] / grp_StoreID['Total Visits'] 
```

```python
grp_StoreID
```

```python
# Grouping by SKUID
```

```python
grp_SkuID = df.groupby('SKU ID')[['Total Visits', 'Transactions']].sum()
grp_SkuID['conversion rate SKU level'] = grp_SkuID['Transactions'] / grp_SkuID['Total Visits'] 
```

```python
grp_SkuID
```

# TASK 2

```python
top_stores = grp_StoreID.sort_values(by='conversion rate Store level', ascending=False).head(5)
top_stores
```

```python
bottom_stores = grp_StoreID.sort_values(by='conversion rate Store level', ascending=True).head(5)
bottom_stores
```

```python
top_skus = grp_SkuID.sort_values(by='conversion rate SKU level', ascending=False).head(5)
top_skus
```

```python
bottom_skus = grp_SkuID.sort_values(by='conversion rate SKU level', ascending=True).head(5)
bottom_skus
```

```python
import matplotlib.pyplot as plt
```

```python
# Plot store-level conversion rates
grp_StoreID['conversion rate Store level'].sort_values().plot(kind='barh', figsize=(10, 6), title='Store Conversion Rates')
plt.xlabel("Conversion Rate")
plt.show()
```

```python

plt.figure(figsize=(8, 8))
plt.pie(grp_StoreID['conversion rate Store level'], 
        labels=grp_StoreID.index,
        autopct='%1.1f%%', 
        startangle=140,
        colors=plt.cm.Paired.colors)

plt.title('Store Conversion Rate Distribution')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
grp_SkuID['conversion rate SKU level'].sort_values().plot(kind='barh', figsize=(10, 6), title='SKU Conversion Rates')
plt.xlabel("Conversion Rate")
plt.show()
```

```python

plt.figure(figsize=(8, 8))
plt.pie(grp_SkuID['conversion rate SKU level'], 
        labels=grp_SkuID.index,
        autopct='%1.1f%%', 
        startangle=140,
        colors=plt.cm.Paired.colors)

plt.title('SKU Conversion Rate Distribution')
plt.axis('equal')
plt.tight_layout()
plt.show()

```
